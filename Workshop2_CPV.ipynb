```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Load the image

```python
img = cv2.imread("images.jfif")
plt.imshow(img)
```

### Using median filter to remove salt and pepper noise

```python
def median_filter(img, kernel_size=3):
    """
    Applies median filter to the input image to remove salt and pepper noise.

    Parameters:
        img (np.array): The input image.
        kernel_size (int, optional): The size of the median filter's kernel. Defaults to 3.
    
    Returns:
        np.array: The filtered image.
    """

    # Pad the input image with zeros around its edges to ensure the median filter can be applied to the borders
    padded_img = np.pad(img, [(kernel_size - 1) // 2, (kernel_size - 1) // 2], mode="edge")
    
    # Initialize the filtered image with zeros, with the same shape as the input image
    filtered_img = np.zeros_like(img)
    
    # Iterate over each pixel in the input image
    for row in range(img.shape[0]):
        for col in range(img.shape[1]):
            # Extract the kernel from the padded image
            window = padded_img[row:row + kernel_size, col:col + kernel_size]
            
            # Calculate the median of all the pixels within the kernel
            filtered_img[row, col] = np.median(window)
    
    # Return the filtered image
    return filtered_img
```

```python
filter_img = median_filter(img= img)
plt.imshow(filter_img)
```
